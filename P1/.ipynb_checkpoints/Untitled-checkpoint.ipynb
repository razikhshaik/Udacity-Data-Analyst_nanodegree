Question 3: Report some descriptive statistics regarding this dataset. Include at least one measure of central tendency and at least one measure of variability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from math import sqrt

stroop_data = pd.read_csv("stroopdata.csv")


incongruent_mean = np.mean(stroop_data.Incongruent)
incongruent_median = np.median(stroop_data.Incongruent)
sum_incongruent_variance = (stroop_data.Incongruent-incongruent_mean)*(stroop_data.Incongruent-incongruent_mean)
incongruent_variance = sum_incongruent_variance.sum()/(stroop_data.Incongruent.size - 1)
incongruent_sd = sqrt(incongruent_variance)

congruent_mean = np.mean(stroop_data.Congruent)
congruent_median = np.median(stroop_data.Congruent)
sum_congruent_variance = (stroop_data.Congruent-congruent_mean)*(stroop_data.Congruent-congruent_mean)
congruent_variance = sum_congruent_variance.sum()/(stroop_data.Congruent.size - 1)
congruent_sd = sqrt(congruent_variance)


print "Congruent sample statistics :"
print "mean: " + str(congruent_mean)
print "median: " +  str(congruent_median)
print "variance: " + str(congruent_variance)
print "standard deviation: " + str(congruent_sd)
print "---------------------------------"
print "Incongruent sample statistics :"
print "mean: " + str(incongruent_mean)
print "median: " +  str(incongruent_median)
print "variance: " + str(incongruent_variance)
print "standard deviation: " + str(incongruent_sd)
```

Question 4: Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

```python
%matplotlib inline

stroop_data.plot.box()


#Polar Chart


r = stroop_data.Congruent
r = r.append(pd.Series(stroop_data.Congruent[0]))
theta = np.linspace(0, 2*np.pi, r.size)

r1 = stroop_data.Incongruent
r1 = r1.append(pd.Series(stroop_data.Incongruent[0]))
theta1 = np.linspace(0, 2*np.pi, r1.size)


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='polar')

ax.plot(theta, r, color='b', linewidth=3,ls = '-', label = 'Congruent data')
ax.plot(theta1, r1, color='r', linewidth=3,ls = '-', label = 'Incongruent data')
ax.set_rmax(max(max(r),max(r1)))
ax.set_thetagrids(np.arange(0, 360, 360.0/r.size), labels=[i for i in range(1,r.size)])
ax.grid(True)

ax.legend(loc="lower left")   
ax.set_title("Polar Radar graph", va='bottom')

plt.show()
```

Question 5: Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

```python
print "sample size: " + str(stroop_data.Congruent.size)
difference_samples = stroop_data.Congruent - stroop_data.Incongruent
difference_samples_mean = difference_samples.mean() 
squared_difference_sample = (difference_samples- difference_samples_mean) * (difference_samples- difference_samples_mean)
sd_difference_sample = sqrt(squared_difference_sample.sum()/(stroop_data.Congruent.size - 1))

print "Standard deviation for difference sample: "+ str(sd_difference_sample)

t_statistic = difference_samples_mean/(sd_difference_sample/sqrt(stroop_data.Congruent.size))
print "t-statistic: "+ str(t_statistic)
```
